```
import requests
from bs4 import BeautifulSoup as bs
```

```
base_url = 'https://editorial.rottentomatoes.com/guide/140-essential-action-movies-to-watch-now/2/'
```

```
r = requests.get(base_url)
r.status_code
```

```
html = r.content
soup = bs(html, 'lxml')
```

```
print(soup.prettify())
```

```
# using the following code to invoke a browser download of the file: https://colab.research.google.com/notebooks/io.ipynb#scrollTo=eikfzi8ZT_rW
from google.colab import files

with open("Rotten_tomatoes_page_2_lxml_parser.html", 'wb') as file:
  file.write(soup.prettify('utf-8'))

files.download('Rotten_tomatoes_page_2_lxml_parser.html')

# we can then open this file in Visual Studios and inspect the html code to compare it directly to the actual webpage (to make sure it was parsed correctly, etc)
```

```
# by searching through the html & inspector tool we see that all the info we need for each movie is contained inside a div with class 'col-sm-18':
info_box = soup.find_all("div", {"class": "col-sm-18 col-full-xs countdown-item-content"})
```

```
len(info_box)
```

**Extract the Title, Year & Score for Each Movie:**

```
# extracting all 'h2' tags
headings = [info.find('h2') for info in info_box]
# note that you have to use 'find' rather than 'find_all', bc 'find_all' returns a list of lists vs 'find' returns one list
headings
```

```
headings[0]
```

```
# extract the movie name
movie_names = [heading.find('a').string for heading in headings]
# if you'd used 'find_all' in creating the headings list, then you'd have a list of lists and you'd need to do a 'for' loop to get each list 
movie_names
```

```
# extract the release year
years = [heading.find('span', {'class': 'start-year'}).string for heading in headings]
years
```

```
# strip the parentheses from the years and convert the string to an integer:
years = [int(year.strip('()')) for year in years]
years
```

```
# extract the **Score**
scores = [heading.find('span', {'class': 'tMeterScore'}).string for heading in headings]
scores
```

```
scores = [int(score.strip("%")) for score in scores]
scores
```

```
# count the number of movies that were rated 100%...
hundos = scores.count(100)
print(f"There were {hundos} movies rated 100% on this list from Rotten Tomatoes...")
```

**Extracting Other Info: Critics Consensus; Synopsis; Starring; Directed By:**

```
# extracting all of the critics consensus
consensus = [info.find('div', {'class': 'critics-consensus'}) for info in info_box]
consensus
```

**There are two ways to extract just the Consensus (i.e. without the opening "Critics Consensus: ")**

**Method #1: Text Manipulation**

```
# extracting all critics consensus text
[con.text for con in consensus]
```

```
# we can see that all of the text appears with the term 'Critics Consensus: '
# let's create a string out of this and count the number of characters in it:
open_string = 'Critics Consensus: '
open_len = len(open_string)
open_len
```

```
# this method cuts out the opening 19 characters (i.e. the number of characters in the string: 'Critics Consensus'), no matter what they are:
clean_cons = [con.text[open_len:] for con in consensus]
clean_cons
```

```
# however, there could be an exception on the webpage where the opening phrase isn't there for some reason, so let's adjust for that for some safety:
clean_cons = [con.text[open_len:] if con.text.startswith(open_string) else con.text for con in consensus]
clean_cons
```

**Method 2: Inspecting the HTML**

```
consensus[0]
# by inspecting the first consensus list element, we see that there is a span element that encapsulates the common phrase (i.e. Critics Consensus), and the text follows that Span element
```

```
# we use the contents attribute to obtain the second child, giving us the desired text
consensus[0].contents[1]
```

```
# inspecting the second element we see that the string begins with a space
# so let's strip the spaces at the beginning & end of the text
consensus[0].contents[1].strip()
```

```
# let's do this for all the elements of the consensus list:
consensus_text = [con.contents[1].strip() for con in consensus]
consensus_text
```

**Extracting the Directors:**

```
directors = [info.find('div', {'class': 'info director'}) for info in info_box]
# note that this works whether you use search for class with attribute 'info' or attribute 'director' or attribute 'info director' etc
len(directors)
```

```
directors[0]
```

```
# we can see that the Director's name is the text within an <a> tag, so we can just do a regular list comprehension for this
# however, after running the following lines of code, we see that there are some missing links which throws off an error message saying:
# 'NoneType' object has no attribute 'string'
# director_list = [director.find('a').string for director in directors]
# director_list
# this doesn't work because there are missing links, which we can see by running the following code and then inspecting the output:
director_list = [director.find('a') for director in directors]
director_list
```

```
director_list = [None if director.find('a') is None else director.find('a').string for director in directors]
director_list
```

**Extracting the Cast info**

```
cast_info = [info.find('div', {'class': 'info cast'}) for info in info_box]
cast_info
```

```
# as always, we inspect the first item in the list to understand each element's structure:
cast_info[0]
```

```
len(cast_info)
```

```
# we see that each cast member is represented by a link, and there are multiple cast members for each movie
# for the first element we'll get all the links:
cast_links = cast_info[0].find_all('a')
cast_links
```

```
cast_names = [c.string for c in cast_links]
cast_names
```

```
# we need to merge these into one string, lest we have a list of lists (i.e. a 2D list, which is perfetly fine, but just more complex to process):
cast = ", ".join(cast_names)
cast
```

**We'll now apply these operations to all movies using two methods: (1) a For loop, and (2) list comprehension**

```
# method 1: good ol' for loop:
# initialize the currently empty list of cast members:
cast = []
# loop over every movie
for c in cast_info:
  cast_links = c.find_all('a')
  cast_names = [link.string for link in cast_links]
  result = ", ".join(cast_names)

  cast.append(result)

cast
```

```
# method 2: nested list comprehension:
cast = [", ".join([link.string for link in c.find_all('a')]) for c in cast_info]
cast
```

**Extracting the Movie Synopsis**

```
syn = [info.find('div', {'class': 'info synopsis'}) for info in info_box]
syn
```

```
# inspect the first element
syn[0]
```

```
syn[0].contents[1].strip()
```

```
synopsis_list = [s.contents[1].strip() for s in syn]
synopsis_list
```

**Extracting the Hidden ('Adjusted') Score:**

```
blanks = [info.find('div', {'class': 'row countdown-item-details'}) for info in info_box]
blanks
```

```
blanks[0]
```

```
adj_scores = [blank.find('div', {'class': 'info countdown-adjusted-score'}) for blank in blanks]
adj_scores
```

```
# By inspection we see that the string we are looking for is the second child of the 'div' tag
# adj_scores[0].contents[0].string       #this line gives us the string 'Adjusted Score: '
adj_scores[0].contents[1].strip()
```

```
adj_adj_scores = [score.contents[1].strip() for score in adj_scores]
adj_adj_scores
```

```
adj_adj_scores = [float(score.strip("%")) for score in adj_adj_scores]
adj_adj_scores
```

**Representing the Data in Structured Form:**

```
import pandas as pd
# a dataframe is a data type for storing tabular data
```

```
movies_info = pd.DataFrame()

movies_info["Movie_Title"] = movie_names
movies_info["Year"] = years
movies_info["Score"] = scores
movies_info["Adjusted_Score"] = adj_adj_scores
movies_info["Director"] = director_list
movies_info["Synopsis"] = synopsis_list
movies_info["Cast"] = cast
movies_info["Consensus"] = consensus_text

movies_info
```

```
# here we're just changing the settings so that it doesn't cut off/truncate any of the text displayed in the dataframe:
pd.set_option('display.max_colwidth', -1)
movies_info
```

```
# export the table to a csv & an excel file:
movies_info.to_csv("movies_info.csv", index = False, header = True)
movies_info.to_excel("movies_info.xlsx", index = False, header = True)

# for both of these methods we can see that:
# the first element is the name of the file that we're creating
# the second and third elements are saying to write the index & column names into the file
```
